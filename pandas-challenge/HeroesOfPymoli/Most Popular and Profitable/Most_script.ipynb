```python
import pandas as pd

game_csv = "Resources/purchase_data.csv"


purchase_data = pd.read_csv(game_csv)
purchase_data.head()
```

```python
items = purchase_data[["Item ID", "Item Name", "Price"]]

item_data = items.groupby(["Item ID","Item Name"])

item_count = item_data["Price"].count()

item_value = round(item_data["Price"].sum(),2) 

item_price = round(item_value/item_count,2)

popular_item_df = pd.DataFrame({"Purchase Count": item_count, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":item_value})

most_pop = popular_item_df.sort_values(["Purchase Count"], ascending=False)
most_pop.head()
```
