```python
import pandas as pd

game_csv = "Resources/purchase_data.csv"


purchase_data = pd.read_csv(game_csv)
purchase_data.head()
```

```python
final_player_count = len(purchase_data["SN"].unique())

gender_data = purchase_data.groupby("Gender")

gender_count = gender_data.nunique()["SN"]

player_pct = gender_count / final_player_count *100
player_pct = round(player_pct,2)
```

```python
gender_table = pd.DataFrame({"Percentage of Players": player_pct,"Total Count": gender_count})
gender_table
```
